# Preparar los datos

```python
# Paths de los datos

# Dataset completo (no funciona)
# annotations_filepath =  folder/'TRAINING_dataset'/'training_dataset_json_like_coco_annotations'/'whole_dataset.json'
# img_dir =  folder/'TRAINING_dataset'/'whole_training_dataset'/'img'

# Dataset sin Imagen_129
annotations_filepath = folder/'Dataset'/'whole_dataset_without_129.json'    # <- acá va el path al json estilo coco
img_dir = folder/'TRAINING_dataset'/'whole_training_dataset'/'img'          # <- acá va el path a la carpeta con los jpg
```

```python
# Cargar los datos
parser = parsers.COCOMaskParser(annotations_filepath=annotations_filepath, img_dir=img_dir)
```

```python
# Dividir los datos
data_splitter = RandomSplitter(probs=[0.75,0.15,0.10], seed=18)  
train_records, valid_records, test_records = parser.parse(data_splitter)
```

```python
# Transformaciones (resize y data augmentation)
image_size = (814, 384)
train_tfms = tfms.A.Adapter([*tfms.A.aug_tfms(size=image_size, presize=None, shift_scale_rotate=None, rgb_shift=None, blur=None, crop_fn=None), tfms.A.Normalize()])
valid_tfms = tfms.A.Adapter([*tfms.A.resize_and_pad(image_size), tfms.A.Normalize()])
test_tfms = tfms.A.Adapter([*tfms.A.resize_and_pad(image_size), tfms.A.Normalize()])
```

```python
# Crear los datasets
train_ds = Dataset(train_records, train_tfms)
valid_ds = Dataset(valid_records, valid_tfms)
test_ds = Dataset(test_records, test_tfms)
```

```python
# Mostrar diferentes versiones de una misma imagen
samples = [train_ds[0] for _ in range(3)]
show_samples(samples, ncols=3)
```

# Mask R-CNN

## Preparar el modelo (Icevision)

```python
# Elegir el modelo y el backbone
model_type = models.mmdet.mask_rcnn
backbone = model_type.backbones.resnet50_fpn_1x
```

```python
# Descargar el modelo pre-entrenado
model = model_type.model(backbone=backbone(pretrained=True), num_classes=len(parser.class_map)) 
```

```python
# Crear los dataLoaders
train_dl = model_type.train_dl(train_ds, batch_size=8, num_workers=2, shuffle=True)
valid_dl = model_type.valid_dl(valid_ds, batch_size=8, num_workers=2, shuffle=False)
infer_dl = model_type.infer_dl(valid_ds, batch_size=1, shuffle=False)
```

## Entrenamiento (FastAI)

```python
# Definir las métricas
metrics = [COCOMetric(metric_type=COCOMetricType.mask)]

# Cargar el learner de FastAI
learn = model_type.fastai.learner(dls=[train_dl, valid_dl], model=model, metrics=metrics)
```

```python
# Buscar un learning rate apropiado
learn.lr_find()
```

```python
# Entrenar el modelo
learn.fine_tune(10, 5e-4, freeze_epochs=2)
```
